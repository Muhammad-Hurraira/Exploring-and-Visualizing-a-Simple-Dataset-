```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
# We can load it directly from seaborn
iris = sns.load_dataset('iris')
```

```python
print("Dataset Shape:")
iris.shape
```

```python
print("\nDataset Columns:")
iris.columns
```

```python
iris.head()
```

```python
# 1. Scatter plot to analyze relationships between variables
plt.figure(figsize=(12, 6))
plt.suptitle("Scatter Plots of Iris Dataset Features")

plt.subplot(1, 2, 1)
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Sepal Width")

plt.subplot(1, 2, 2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Petal Width")

plt.tight_layout()
plt.show()
```

```python
# 2. Histograms to examine data distribution
plt.figure(figsize=(12, 8))
plt.suptitle("Distribution of Iris Dataset Features")

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=iris, x=feature, hue='species', kde=True, element='step')
    plt.title(f'Distribution of {feature.replace("_", " ").title()}')

plt.tight_layout()
plt.show()
```

```python
# 3. Box plots to detect outliers and spread of values
plt.figure(figsize=(12, 6))
plt.suptitle("Box Plots of Iris Dataset Features")

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='species', y=feature)
    plt.title(f'Box Plot of {feature.replace("_", " ").title()}')

plt.tight_layout()
plt.show()
```

```python
# Bonus: Pairplot for comprehensive visualization
print("\nGenerating pairplot (this may take a moment)...")
sns.pairplot(iris, hue='species')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```
